```python
import pandas as pd
```

```python
#Let's take a look at this data!

df = pd.read_csv('pz_test_run_data_2.csv')
```

```python
print(df.shape)
print(df.info())
df.head()
```

```python
# Pivot to go from a bunch of key value paris to having data about each unique ein/period combo
variables_pivoted = df.pivot_table(index = ['ein', 'period'], columns = 'variable', values = 'value', aggfunc='max')
print(variables_pivoted.shape)
variables_pivoted.head()
```

```python
#Let's get some names that are a little easier to understand

import pickle

with open('name_mapper_2.pkl', 'rb') as picklefile:
    name_mapper = pickle.load(picklefile)
```

```python
variables_pivoted.rename(name_mapper, axis='columns', inplace=True)
variables_pivoted.head()
```

```python
variables_pivoted.isna().sum().sort_values(ascending=False)
```

```python
variables_pivoted[variables_pivoted.Inventory_Sales.isna()].merge(
    variables_pivoted[variables_pivoted.Inventory_Sales_2.isna()], how='inner').shape
```

```python
variables_pivoted[variables_pivoted.Cost_of_Goods_Sold.isna()].merge(
    variables_pivoted[variables_pivoted.Cost_of_Goods_Sold_2.isna()], how='inner').shape
```

```python
#These variables don't seem to have enough data to be useful

variables_pivoted.drop(['Inventory_Sales', 'Inventory_Sales_2', 'Cost_of_Goods_Sold', 'Cost_of_Goods_Sold_2',
                        'Rev_Special_Events_2', 'Special_Events_Net', 'Special_Events_Exp', 'Membership_Dues'], axis=1, inplace=True)
variables_pivoted.head()
```

```python
variables_pivoted['Excess_Deficit_1'].fillna(variables_pivoted['Excess_Deficit_2'], inplace=True)
variables_pivoted.drop('Excess_Deficit_1', axis=1, inplace=True)
```

```python
variables_pivoted['Employee_Comp_2'].fillna(variables_pivoted['Employee_Comp_1'], inplace=True)
variables_pivoted.drop('Employee_Comp_1', axis=1, inplace=True)
```
